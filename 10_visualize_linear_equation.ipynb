```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
y1 = 3 * x
```

```python
y2 = 1 + (5*x)/2
```

```python
# create an empty plot
fig, ax = plt.subplots()
# label x axis
plt.xlabel("x")
# label y axis
plt.ylabel("y")
# set limit to x axis
ax.set_xlim([0,3])
# set limit to y axis
ax.set_ylim([0,8])
# input of x will give us y for y1 equation
ax.plot(x, y1, c="green")
# input of x will give us y for y2 equation
ax.plot(x, y2, c="red")
# plot the point of intersection vertically
plt.axvline(x=2, color="purple", linestyle='--')
# plot the point of intersection horizontally
plt.axhline(y=6, color="purple", linestyle='--')

```
